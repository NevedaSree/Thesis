```python
import sunau
import os
import pylab
import wave


genres = ["blues","classical","country","disco","hiphop","jazz","metal","pop","reggae","rock"]

def graph_spectrogram(wav_file,i,j,name):
    sound_info, frame_rate = get_info(wav_file,name)
    #print(sound_info)
    pylab.figure(num=None, figsize=(19, 12))
    pylab.subplot(111)

    pylab.specgram(sound_info, Fs=frame_rate)
    pylab.savefig('C:/Users/neved/Desktop/spectrogram/music'+str(j)+str(i)+'.png')
    pylab.close()


def get_info(wav_file,name):
    file= wave.open("C:/Users/neved/Desktop/wavefile/"+name+"/"+wav_file, 'r')
    frames = file.readframes(file.getnframes())
    print(file.getparams())
    print(frames)
    sound_info = pylab.fromstring(frames, 'Int16')
    frame_rate = file.getframerate()
    file.close()
    return sound_info, frame_rate


def loadData():
    
    bluesList = os.listdir('C:/Users/neved/Desktop/wavefile/blues')
    classicalList = os.listdir('C:/Users/neved/Desktop/wavefile/classical')
    countryList = os.listdir('C:/Users/neved/Desktop/wavefile/country')
    discoList = os.listdir('C:/Users/neved/Desktop/wavefile/disco')
    hiphopList = os.listdir('C:/Users/neved/Desktop/wavefile/hiphop')
    jazzList = os.listdir('C:/Users/neved/Desktop/wavefile/jazz')
    metalList = os.listdir('C:/Users/neved/Desktop/wavefile/metal')
    popList = os.listdir('C:/Users/neved/Desktop/wavefile/pop')
    reggaeList = os.listdir('C:/Users/neved/Desktop/wavefile/reggae')
    rockList = os.listdir('C:/Users/neved/Desktop/wavefile/rock')

    listAll = [bluesList,classicalList,countryList,discoList,hiphopList,jazzList,metalList,popList,reggaeList,rockList]

    j=0

    for genre in listAll:
        i=0
        if j == 0:
            name = "blues"
        elif j == 1:
            name = "classical"
        elif j == 2:
            name = "country"
        elif j == 3:
            name = "disco"
        elif j == 4:
            name = "hiphop"
        elif j == 5:
            name = "jazz"
        elif j == 6:
            name = "metal"
        elif j == 7:
            name = "pop"
        elif j == 8:
            name = "reggae"
        elif j == 9:
            name = "rock"
        for song in listAll[j]:
            # print(song)
            graph_spectrogram(song,i,j,name)
            i=i+1
        j=j+1


if __name__ == '__main__':
    loadData()
```
